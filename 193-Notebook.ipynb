```python
import pandas as pd
```

```python
# Les urls des données se trouvent...
```

```python
URLS_DATA = pd.read_csv('http://www.aufilduboamp.com/DOCS/CAHIER_D_ESSAIS_ANNEXES/urls_aufilduboamp_boamp_docs.csv', usecols=["urls_aufilduboamp_boamp_docs"])
```

```python
# Il convient d'en tirer une liste issue de la colonne...
```

```python
URLS_DATA = URLS_DATA.urls_aufilduboamp_boamp_docs.to_list()
```

```python
# Dans cette liste, sélectionner les occurences de la table D_01
```

```python
liste_acheteurs = [x for x in URLS_DATA if "D_01" in x if "2021" in x]
liste_acheteurs
```

```python
# .................
```

```python
df_BOAMP_2015_2022_ACHETEURS = pd.concat([pd.read_csv(x, usecols=["D_01_01_Identite_Denomination", "BOAMP_AVIS_NUM"])
                                          for x in liste_acheteurs[:]], ignore_index=True)
```

```python
df_BOAMP_2015_2022_ACHETEURS.set_index("BOAMP_AVIS_NUM", inplace=True, drop=True )
```
