```python
import pandas as pd
```

```python
URLS_DATA = pd.read_csv('http://www.aufilduboamp.com/DOCS/urls_aufilduboamp_boamp_docs.csv', usecols=["urls_aufilduboamp_boamp_docs"])
```

```python
URLS_DATA = URLS_DATA.urls_aufilduboamp_boamp_docs.to_list()
```

```python
liste_acheteurs = [x for x in URLS_DATA if "D_01" in x]
```

```python
df_BOAMP_2015_2022_ACHETEURS = pd.concat([pd.read_csv(x, usecols=["D_01_01_Identite_Denomination", "BOAMP_AVIS_NUM"])
                                          for x in liste_acheteurs[:]], ignore_index=True)
```

```python
df_BOAMP_2015_2022_ACHETEURS.set_index("BOAMP_AVIS_NUM", inplace=True, drop=True )
```
